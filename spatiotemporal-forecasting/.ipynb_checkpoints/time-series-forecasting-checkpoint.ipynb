## ğŸ“¦ å¯¼å…¥å¿…è¦çš„åº“


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# è®¾ç½®éšæœºç§å­
torch.manual_seed(42)
np.random.seed(42)

# è®¾ç½®è®¾å¤‡
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ä½¿ç”¨è®¾å¤‡: {device}")

# è®¾ç½®matplotlibä¸­æ–‡å­—ä½“
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## ğŸ“Š æ•°æ®å‡†å¤‡

### ç”Ÿæˆæ¨¡æ‹Ÿæ—¶åºæ•°æ®

ç”±äºETTæ•°æ®é›†è¾ƒå¤§ï¼Œæˆ‘ä»¬å…ˆç”Ÿæˆä¸€äº›æ¨¡æ‹Ÿçš„å¤šå˜é‡æ—¶åºæ•°æ®æ¥æ¼”ç¤ºæ¨¡å‹ï¼š


```python
def generate_synthetic_timeseries(n_samples=1000, n_features=7, seq_len=96, pred_len=24):
    """
    ç”Ÿæˆæ¨¡æ‹Ÿçš„å¤šå˜é‡æ—¶åºæ•°æ®
    
    Args:
        n_samples: æ ·æœ¬æ•°é‡
        n_features: ç‰¹å¾æ•°é‡
        seq_len: è¾“å…¥åºåˆ—é•¿åº¦
        pred_len: é¢„æµ‹åºåˆ—é•¿åº¦
    
    Returns:
        data: æ¨¡æ‹Ÿæ—¶åºæ•°æ®
    """
    np.random.seed(42)
    
    # ç”ŸæˆåŸºç¡€è¶‹åŠ¿å’Œå­£èŠ‚æ€§
    t = np.linspace(0, 10, n_samples)
    
    # ä¸ºæ¯ä¸ªç‰¹å¾ç”Ÿæˆä¸åŒçš„æ¨¡å¼
    data = np.zeros((n_samples, n_features))
    
    for i in range(n_features):
        # è¶‹åŠ¿é¡¹
        trend = 0.1 * t * np.sin(0.5 * t + i)
        
        # å­£èŠ‚æ€§é¡¹
        seasonal = 2 * np.sin(2 * np.pi * t / 24 + i) + np.sin(2 * np.pi * t / 168 + i)
        
        # å™ªå£°é¡¹
        noise = 0.5 * np.random.randn(n_samples)
        
        # ç»„åˆ
        data[:, i] = trend + seasonal + noise
    
    # æ ‡å‡†åŒ–
    scaler = StandardScaler()
    data = scaler.fit_transform(data)
    
    return data, scaler

# ç”Ÿæˆæ•°æ®
data, scaler = generate_synthetic_timeseries(n_samples=1000, n_features=7)
print(f"æ•°æ®å½¢çŠ¶: {data.shape}")

# å¯è§†åŒ–æ•°æ®
plt.figure(figsize=(15, 10))
for i in range(7):
    plt.subplot(3, 3, i+1)
    plt.plot(data[:200, i])
    plt.title(f'ç‰¹å¾ {i+1}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### åˆ›å»ºæ•°æ®é›†ç±»


```python
class TimeSeriesDataset(Dataset):
    """æ—¶åºé¢„æµ‹æ•°æ®é›†ç±»"""
    
    def __init__(self, data, seq_len=96, pred_len=24):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len
        
    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len + 1
    
    def __getitem__(self, idx):
        # è¾“å…¥åºåˆ—
        x = self.data[idx:idx + self.seq_len]
        # ç›®æ ‡åºåˆ—
        y = self.data[idx + self.seq_len:idx + self.seq_len + self.pred_len]
        
        return torch.FloatTensor(x), torch.FloatTensor(y)

# åˆ›å»ºæ•°æ®é›†
seq_len = 96  # è¾“å…¥åºåˆ—é•¿åº¦
pred_len = 24  # é¢„æµ‹é•¿åº¦

dataset = TimeSeriesDataset(data, seq_len=seq_len, pred_len=pred_len)
print(f"æ•°æ®é›†å¤§å°: {len(dataset)}")

# åˆ’åˆ†è®­ç»ƒã€éªŒè¯ã€æµ‹è¯•é›†
train_size = int(0.7 * len(dataset))
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_dataset = torch.utils.data.Subset(dataset, range(train_size))
val_dataset = torch.utils.data.Subset(dataset, range(train_size, train_size + val_size))
test_dataset = torch.utils.data.Subset(dataset, range(train_size + val_size, len(dataset)))

print(f"è®­ç»ƒé›†å¤§å°: {len(train_dataset)}")
print(f"éªŒè¯é›†å¤§å°: {len(val_dataset)}")
print(f"æµ‹è¯•é›†å¤§å°: {len(test_dataset)}")

# åˆ›å»ºæ•°æ®åŠ è½½å™¨
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

## ğŸ§  æ¨¡å‹å®ç°

### 1. Transformeræ—¶åºé¢„æµ‹æ¨¡å‹


```python
class PositionalEncoding(nn.Module):
    """ä½ç½®ç¼–ç """
    
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:x.size(0), :]

class TransformerForecasting(nn.Module):
    """åŸºäºTransformerçš„æ—¶åºé¢„æµ‹æ¨¡å‹"""
    
    def __init__(self, input_dim, d_model=512, nhead=8, num_layers=6, pred_len=24):
        super(TransformerForecasting, self).__init__()
        self.d_model = d_model
        self.pred_len = pred_len
        
        # è¾“å…¥æŠ•å½±å±‚
        self.input_projection = nn.Linear(input_dim, d_model)
        
        # ä½ç½®ç¼–ç 
        self.pos_encoding = PositionalEncoding(d_model)
        
        # Transformerç¼–ç å™¨
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=2048,
            dropout=0.1,
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # è¾“å‡ºæŠ•å½±å±‚
        self.output_projection = nn.Linear(d_model, input_dim)
        
        # é¢„æµ‹å±‚
        self.predictor = nn.Sequential(
            nn.Linear(d_model, d_model // 2),
            nn.ReLU(),
            nn.Linear(d_model // 2, input_dim)
        )
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_dim)
        batch_size, seq_len, input_dim = x.shape
        
        # è¾“å…¥æŠ•å½±
        x = self.input_projection(x)  # (batch_size, seq_len, d_model)
        
        # ä½ç½®ç¼–ç 
        x = x.transpose(0, 1)  # (seq_len, batch_size, d_model)
        x = self.pos_encoding(x)
        x = x.transpose(0, 1)  # (batch_size, seq_len, d_model)
        
        # Transformerç¼–ç 
        encoded = self.transformer_encoder(x)  # (batch_size, seq_len, d_model)
        
        # ä½¿ç”¨æœ€åä¸€ä¸ªæ—¶é—´æ­¥çš„ç¼–ç è¿›è¡Œé¢„æµ‹
        last_hidden = encoded[:, -1, :]  # (batch_size, d_model)
        
        # é¢„æµ‹æœªæ¥åºåˆ—
        predictions = []
        current_input = last_hidden
        
        for _ in range(self.pred_len):
            pred = self.predictor(current_input)  # (batch_size, input_dim)
            predictions.append(pred)
            
            # å°†é¢„æµ‹ç»“æœæŠ•å½±å›d_modelç©ºé—´ä½œä¸ºä¸‹ä¸€æ­¥çš„è¾“å…¥
            current_input = self.input_projection(pred)
        
        # å †å é¢„æµ‹ç»“æœ
        output = torch.stack(predictions, dim=1)  # (batch_size, pred_len, input_dim)
        
        return output

# æµ‹è¯•æ¨¡å‹
model = TransformerForecasting(input_dim=7, d_model=256, nhead=8, num_layers=4, pred_len=pred_len)
model = model.to(device)

# æµ‹è¯•å‰å‘ä¼ æ’­
sample_input = torch.randn(2, seq_len, 7).to(device)
sample_output = model(sample_input)
print(f"è¾“å…¥å½¢çŠ¶: {sample_input.shape}")
print(f"è¾“å‡ºå½¢çŠ¶: {sample_output.shape}")

# è®¡ç®—å‚æ•°é‡
total_params = sum(p.numel() for p in model.parameters())
print(f"æ¨¡å‹å‚æ•°é‡: {total_params:,}")
```

### 2. LSTMæ—¶åºé¢„æµ‹æ¨¡å‹


```python
class LSTMForecasting(nn.Module):
    """åŸºäºLSTMçš„æ—¶åºé¢„æµ‹æ¨¡å‹"""
    
    def __init__(self, input_dim, hidden_dim=128, num_layers=2, pred_len=24, dropout=0.1):
        super(LSTMForecasting, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.pred_len = pred_len
        
        # LSTMå±‚
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0
        )
        
        # é¢„æµ‹å±‚
        self.predictor = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim // 2, input_dim)
        )
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_dim)
        batch_size = x.size(0)
        
        # LSTMå‰å‘ä¼ æ’­
        lstm_out, (hidden, cell) = self.lstm(x)
        
        # ä½¿ç”¨æœ€åä¸€ä¸ªæ—¶é—´æ­¥çš„è¾“å‡º
        last_output = lstm_out[:, -1, :]  # (batch_size, hidden_dim)
        
        # é¢„æµ‹æœªæ¥åºåˆ—
        predictions = []
        current_hidden = hidden
        current_cell = cell
        
        for _ in range(self.pred_len):
            # é¢„æµ‹å½“å‰æ—¶é—´æ­¥
            pred = self.predictor(last_output)  # (batch_size, input_dim)
            predictions.append(pred)
            
            # å°†é¢„æµ‹ç»“æœä½œä¸ºä¸‹ä¸€æ­¥çš„è¾“å…¥
            pred_reshaped = pred.unsqueeze(1)  # (batch_size, 1, input_dim)
            
            # ç»§ç»­LSTMå‰å‘ä¼ æ’­
            lstm_out, (current_hidden, current_cell) = self.lstm(pred_reshaped, (current_hidden, current_cell))
            last_output = lstm_out[:, -1, :]  # (batch_size, hidden_dim)
        
        # å †å é¢„æµ‹ç»“æœ
        output = torch.stack(predictions, dim=1)  # (batch_size, pred_len, input_dim)
        
        return output

# åˆ›å»ºLSTMæ¨¡å‹
lstm_model = LSTMForecasting(input_dim=7, hidden_dim=128, num_layers=2, pred_len=pred_len)
lstm_model = lstm_model.to(device)

# æµ‹è¯•LSTMæ¨¡å‹
lstm_output = lstm_model(sample_input)
print(f"LSTMè¾“å‡ºå½¢çŠ¶: {lstm_output.shape}")

# è®¡ç®—å‚æ•°é‡
lstm_params = sum(p.numel() for p in lstm_model.parameters())
print(f"LSTMæ¨¡å‹å‚æ•°é‡: {lstm_params:,}")
```

## ğŸš€ æ¨¡å‹è®­ç»ƒ


```python
def train_model(model, train_loader, val_loader, epochs=50, lr=0.001):
    """è®­ç»ƒæ¨¡å‹"""
    
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)
    
    train_losses = []
    val_losses = []
    
    for epoch in range(epochs):
        # è®­ç»ƒé˜¶æ®µ
        model.train()
        train_loss = 0.0
        
        for batch_x, batch_y in train_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)
            
            optimizer.zero_grad()
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
        
        train_loss /= len(train_loader)
        
        # éªŒè¯é˜¶æ®µ
        model.eval()
        val_loss = 0.0
        
        with torch.no_grad():
            for batch_x, batch_y in val_loader:
                batch_x, batch_y = batch_x.to(device), batch_y.to(device)
                outputs = model(batch_x)
                loss = criterion(outputs, batch_y)
                val_loss += loss.item()
        
        val_loss /= len(val_loader)
        
        # å­¦ä¹ ç‡è°ƒåº¦
        scheduler.step(val_loss)
        
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}')
    
    return train_losses, val_losses

# è®­ç»ƒTransformeræ¨¡å‹
print("å¼€å§‹è®­ç»ƒTransformeræ¨¡å‹...")
transformer_train_losses, transformer_val_losses = train_model(
    model, train_loader, val_loader, epochs=50, lr=0.001
)

print("\nå¼€å§‹è®­ç»ƒLSTMæ¨¡å‹...")
lstm_train_losses, lstm_val_losses = train_model(
    lstm_model, train_loader, val_loader, epochs=50, lr=0.001
)
```

## ğŸ“Š è®­ç»ƒç»“æœå¯è§†åŒ–


```python
# ç»˜åˆ¶è®­ç»ƒæŸå¤±æ›²çº¿
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(transformer_train_losses, label='Transformer Train Loss', alpha=0.8)
plt.plot(transformer_val_losses, label='Transformer Val Loss', alpha=0.8)
plt.title('Transformeræ¨¡å‹è®­ç»ƒæŸå¤±')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(lstm_train_losses, label='LSTM Train Loss', alpha=0.8)
plt.plot(lstm_val_losses, label='LSTM Val Loss', alpha=0.8)
plt.title('LSTMæ¨¡å‹è®­ç»ƒæŸå¤±')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# å¯¹æ¯”ä¸¤ä¸ªæ¨¡å‹çš„æœ€ç»ˆæŸå¤±
print("æœ€ç»ˆè®­ç»ƒæŸå¤±å¯¹æ¯”:")
print(f"Transformer - è®­ç»ƒæŸå¤±: {transformer_train_losses[-1]:.6f}, éªŒè¯æŸå¤±: {transformer_val_losses[-1]:.6f}")
print(f"LSTM - è®­ç»ƒæŸå¤±: {lstm_train_losses[-1]:.6f}, éªŒè¯æŸå¤±: {lstm_val_losses[-1]:.6f}")
```

## ğŸ§ª æ¨¡å‹è¯„ä¼°ä¸é¢„æµ‹å¯è§†åŒ–


```python
def evaluate_model(model, test_loader):
    """è¯„ä¼°æ¨¡å‹æ€§èƒ½"""
    model.eval()
    predictions = []
    targets = []
    
    with torch.no_grad():
        for batch_x, batch_y in test_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)
            outputs = model(batch_x)
            
            predictions.append(outputs.cpu().numpy())
            targets.append(batch_y.cpu().numpy())
    
    predictions = np.concatenate(predictions, axis=0)
    targets = np.concatenate(targets, axis=0)
    
    # è®¡ç®—è¯„ä¼°æŒ‡æ ‡
    mse = mean_squared_error(targets.flatten(), predictions.flatten())
    mae = mean_absolute_error(targets.flatten(), predictions.flatten())
    rmse = np.sqrt(mse)
    
    return predictions, targets, mse, mae, rmse

# è¯„ä¼°ä¸¤ä¸ªæ¨¡å‹
print("è¯„ä¼°Transformeræ¨¡å‹...")
transformer_preds, transformer_targets, transformer_mse, transformer_mae, transformer_rmse = evaluate_model(model, test_loader)

print("è¯„ä¼°LSTMæ¨¡å‹...")
lstm_preds, lstm_targets, lstm_mse, lstm_mae, lstm_rmse = evaluate_model(lstm_model, test_loader)

# æ‰“å°è¯„ä¼°ç»“æœ
print("\n=== æ¨¡å‹æ€§èƒ½å¯¹æ¯” ===")
print(f"{'æ¨¡å‹':<15} {'MSE':<10} {'MAE':<10} {'RMSE':<10}")
print("-" * 45)
print(f"{'Transformer':<15} {transformer_mse:<10.6f} {transformer_mae:<10.6f} {transformer_rmse:<10.6f}")
print(f"{'LSTM':<15} {lstm_mse:<10.6f} {lstm_mae:<10.6f} {lstm_rmse:<10.6f}")
```

```python
# å¯è§†åŒ–é¢„æµ‹ç»“æœ
def plot_predictions(predictions, targets, model_name, sample_idx=0, feature_idx=0):
    """ç»˜åˆ¶é¢„æµ‹ç»“æœå¯¹æ¯”å›¾"""
    plt.figure(figsize=(12, 6))
    
    # é€‰æ‹©ä¸€ä¸ªæ ·æœ¬è¿›è¡Œå¯è§†åŒ–
    pred_sample = predictions[sample_idx, :, feature_idx]
    target_sample = targets[sample_idx, :, feature_idx]
    
    time_steps = range(len(pred_sample))
    
    plt.plot(time_steps, target_sample, 'b-', label='çœŸå®å€¼', linewidth=2)
    plt.plot(time_steps, pred_sample, 'r--', label='é¢„æµ‹å€¼', linewidth=2)
    
    plt.title(f'{model_name} - ç‰¹å¾{feature_idx+1}é¢„æµ‹ç»“æœå¯¹æ¯”')
    plt.xlabel('æ—¶é—´æ­¥')
    plt.ylabel('æ•°å€¼')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# ç»˜åˆ¶Transformeré¢„æµ‹ç»“æœ
plot_predictions(transformer_preds, transformer_targets, 'Transformer', sample_idx=0, feature_idx=0)

# ç»˜åˆ¶LSTMé¢„æµ‹ç»“æœ
plot_predictions(lstm_preds, lstm_targets, 'LSTM', sample_idx=0, feature_idx=0)
```

## ğŸ¯ æ¨¡å‹æ€§èƒ½æ€»ç»“
